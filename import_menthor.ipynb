```python
import os
import datetime
import pandas as pd
import boto3

from botocore.exceptions import ClientError

# create date range descending from today to 6 months ago
# Get the current date
start_date = datetime.datetime(2023, 7, 28)

# Calculate the start date 6 months ago
end_date = datetime.datetime(2023, 5, 1)

# Create the date range descending from today to 6 months ago
date_range = pd.date_range(start=end_date, end=start_date, freq="B")

# Convert the date range to a list
date_range_list = date_range.strftime("%Y-%m-%d").tolist()

# reverse the date_range order
date_range = date_range[::-1]

date_range = ["2023-08-04"]

# Create an S3 client
s3 = boto3.client("s3")

# Specify the S3 bucket
bucket_name = "q-models"

folder_name = "menthor"

# URL of the image to download
image_urls = [
    f"spx/all_exp_profile",
    f"qreport/image+(2)",
    f"qreport/image",
    f"spx/main_chart",
    f"spx/next_exps_profile",
    f"spx/volume_oi_all_exp",
    f"spx/volume_oi_0dte",
    f"spx/styled_table_spx",
    f"spx/pos_net_gex_table_spx",
    f"spx/neg_net_gex_table_spx",
    f"spx/dex_gex_price_past",
    f"spx/0dte_spread",
    f"spx/term_structure2",
    f"spx/risk_reversal",
    f"qqq/dex_gex_price",
    f"qqq/styled_table_qqq",
    f"qqq/all_exp_profile",
    f"qqq/term_structure2",
    f"vix/dex_gex_price",
    f"vix/styled_table_vix",
    f"vix/all_exp_profile",
    f"vix/term_structure2",
]


for dt in date_range_list:
    # create folder with current date
    current_date = dt

    for image_url in image_urls:
        # Extract the filename from the URL
        filename = "_".join(image_url.split("/"))

        # Construct the object name
        object_name = f"{current_date}/{image_url}.png"

        # Local filename to save the downloaded file
        local_file_name = f"menthor/{filename}_{current_date}.png"

        print(f"{current_date} {object_name} -> {local_file_name}")

        if os.path.exists(local_file_name):
            print(f"{local_file_name} already exists. Skipping download.")
        else:
            try:
                # Attempt to download the file
                s3.download_file(bucket_name, object_name, local_file_name)
            except ClientError as e:
                # Check if the error is due to the file not found or access denied
                error_code = e.response["Error"]["Code"]
                if error_code == "404":
                    print("Error: The file was not found in the bucket.")
                elif error_code == "403":
                    print("Error: Access to the file in the bucket is denied.")
                else:
                    print(f"An unexpected error occurred: {e}")
```
